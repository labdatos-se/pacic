```python
import requests
import pandas as pd
```

```python
sucursales_CDMX = pd.read_excel("C:/Users/carlos.ramirezc/Documents/Laboratorio/PACIC/Sucursales Cornershop CDMX.xlsx")
```

```python
sucursales_CDMX
```

```python
#Quitamos duplicados de id_tienda
sucursales_CDMX = sucursales_CDMX.drop_duplicates(subset=['id_tienda'], keep='first')
sucursales_CDMX
```

```python
url_basica="https://cornershopapp.com/api/v3/branches/"
url_fin="?with_suspended_slots&app_capabilities=DATA_REQUEST_BASIC_FORM,CUSTOMER_WITH_SENDBIRD_CHAT,DATA_REQUEST_PRESCRIPTION,LOYALTY_PROVIDERS_WITH_2FA,IMAGE_SUPPORT_CHAT,TC_BLOCKER,CUSTOMER_SMART_SERVICE_FEE,STORE_WITHOUT_PRICING_NOTES"
```

```python
id_sucursal=[]
sucursales=[]
latitud=[]
longitud=[]
store_name=[]
store_id=[]
address=[]

for branches in sucursales_CDMX['id_tienda']:
    url=url_basica+str(branches)+url_fin
    r = requests.get(url)
    data_branches = r.json()
    id_sucursal.append(data_branches['branch']['id'])
    sucursales.append(data_branches['branch']['name'])
    latitud.append(data_branches['branch']['lat'])
    longitud.append(data_branches['branch']['lng'])
    store_name.append(data_branches['branch']['store_name'])
    store_id.append(data_branches['branch']['store_id'])
    address.append(data_branches['branch']['address'])
```

```python
#pasa los datos a un dataframe
df=pd.DataFrame({"id_sucursal":id_sucursal,"sucursal":sucursales,"latitud":latitud,"longitud":longitud,"store_name":store_name,"store_id":store_id,"address":address})
```
